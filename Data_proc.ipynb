```python
# Fixing the size column 
# first we will make a function to convert the size to numeric
def converting_size_to_numeric(size):
    if not isinstance(size, str):
        return np.nan
    
    size = size.strip().upper()
    if size.endswith('M'):
        try:
            return float(size[:-1])
        except ValueError:
            return np.nan
    elif size.endswith('K'):
        try:
            return float(size[:-1]) / 1000
        except ValueError:
            return np.nan
    else:
        try:
            return float(size)
        except ValueError:
            return np.nan

# now we apply it to the size column
play_store['Size'] = play_store['Size'].apply(converting_size_to_numeric)
```

```python
# we wanted to make sure that all the sizes is calculated correctly that is under 1
play_store.rename(columns={'Size':'Size in MB'}, inplace=True) 
play_store[play_store['Size in MB']<1]
 # renaming the size column to size in MB instead of size to make it more clear
play_store['Size in MB'].fillna(play_store['Size in MB'].mean(), inplace=True)  # filling the missing values with the mean
pd.to_numeric(play_store['Size in MB'])  # converting the size in MB to numeric
```

```python
# now we want to fix the prices column
play_store[play_store['Price'] == '']  # wanted if there is any empty values in the price column
def dollars_to_numbers(price):
    if  isinstance(price, str):
        value = price.strip('$')
        return float(value)
    else:
        return price
    

play_store['Price'] = play_store['Price'].apply(dollars_to_numbers)
```

```python
# fixing the android version column and turning it into numeric
string = '4.4 and up'
print(string.find(' and up'))  # just for experminting 
string = string.replace(' and up', '')
print(string)
def android_version_to_numeric(version):
    def convert_to_numeric(v):
        v = v.replace('W', '')
        parts = v.split('.')
        return float(parts[0] + '.' + ''.join(parts[1:]))
    
    if ' and up' in version:
        base_version = version.replace(' and up', '')
        latest_version = float('inf')  # turning the latest version to infinity
        return convert_to_numeric(base_version), latest_version
    elif ' - ' in version:
        base_version, latest_version = version.split(' - ')
        return convert_to_numeric(base_version), convert_to_numeric(latest_version)
    elif version == 'Varies with device':
        return np.nan, np.nan
    elif version == np.nan:
        return np.nan, np.nan
    else:
        return convert_to_numeric(version), convert_to_numeric(version)


play_store[['base_version', 'latest_version']] = play_store['Android Ver'].apply(
    lambda x: pd.Series(android_version_to_numeric(x))
)
```
